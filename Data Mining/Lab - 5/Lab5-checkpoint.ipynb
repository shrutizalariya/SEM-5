### 2)  Handle Missing Values in data set [use dropna(), fillna(), and interpolate]


```python
#delete records(row) which contains null values(not physible)(defaul delete row)
#with dropna
data_withdropna = df.copy()
data_withdropna = data_withdropna.dropna()
data_withdropna
```

```python
# axis=1 so it deletes column which contain 1 or more null value
data_withdropna = df.copy()
data_withdropna = data_withdropna.dropna(how='any',axis=1)
data_withdropna
```

```python
# axis=1 so it deletes columns in which all the values are null 
data_withdropna = df.copy()
data_withdropna = data_withdropna.dropna(how='all',axis=1)
data_withdropna
```

```python
#with fillna
#globally
data_withfillna = df.copy()
data_withfillna = data_withfillna.fillna(1000) #----->insert 1000 where value is null in all numeric columns
data_withfillna
```

```python
#ColumnWise
data_withfillna = df.copy()
data_withfillna = data_withfillna.fillna({'Age':100,'Cabin':'Not Available'})
data_withfillna
```

```python
data_withfillna = df.copy()
meanAge = data_withfillna.Age.mean()
print(meanAge)
data_withfillna = data_withfillna.fillna({'Age':meanAge,'Cabin':'Not Available'})
data_withfillna
```

```python
data_withfillna = df.copy()
meanAge = data_withfillna.Age.mean()
modeCabin = data_withfillna.Cabin.mode()[0] #mode can be multible but we will take first one so that
print(meanAge)
print(modeCabin)
data_withfillna = data_withfillna.fillna({'Age':meanAge,'Cabin':modeCabin})
data_withfillna
```

```python
#interpolate only works on numeric not on strings and objects
#interpolate uses for replace columns null value according their up and down rows value
data_interpolate = df.copy()
data_interpolate = data_interpolate.interpolate()
data_interpolate
```

### 3)  Apply Scaling to AGE attribute with min max, decimal scaling and z score.


```python
#With min-max
data = df.copy()

minAge = int(data.Age.min())
maxAge = int(data.Age.max())

print('Min Age =',minAge) # 0
print('Max Age =',maxAge) # 80

Ages = data.Age

data['MinMaxAge'] = (Ages - minAge) / (maxAge - minAge)
data
```

```python
# with decimal scaling
data = df.copy()
maxAge = data.Age.max()
noOfDigits = len(str(int(maxAge)))
Ages = data.Age
print(noOfDigits)
data['DecimalScalingAge'] = Ages/(10**noOfDigits)
data
```
